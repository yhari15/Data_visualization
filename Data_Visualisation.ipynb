```python
#data visualisation using Matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,20,10)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show() #to remove index or heading
```

```python
plt.plot(x,y,color='g') #--to change color use it 
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Waveform')
plt.show()
```

```python
#lineplot------>1D plot
plt.plot(x,y,color='g',alpha=0.2,linestyle='--',marker='o',label='Wave') #--to change transparency of line use alpha, to change linestyle use it  
#plt.axis([0,30,0,500]) --->to change axis limit
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Waveform')
plt.legend() #--to name a graph
plt.show()
```

```python
#to plot multiple plots
plt.plot(x,np.sin(x),label='Sin_wave')
plt.plot(x,np.cos(x),label='Cos_wave')
plt.legend()
plt.show()
```

```python
#Multiple plot in seperate window
plt.plot(x,np.sin(x),label='Sin_wave')
plt.legend()
plt.show()   #to seperate windows of graph
plt.plot(x,np.cos(x),label='Cos_wave')
plt.legend()
plt.show()
```

```python
#Subplot---------->for 20 graphs the above one would be difficult so we use Subplot
plt.figure(figsize=(15,8)) #----->to fix dimensions of the graph
plt.subplot(2,2,1)
plt.plot(x,np.sin(x),color='r',label='Sin_wave')
plt.legend()
plt.subplot(2,2,2)
plt.plot(x,np.cos(x),color='g',label='Cos_wave')
plt.legend()
plt.subplot(2,2,3)
plt.plot(x,np.tan(x),color='b',label='tan_wave')
plt.legend()
plt.subplot(2,2,4)
plt.plot(x,y,color='y',label='Waveform')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5)) #----->to fix dimensions of the graph
plt.subplot(2,2,1)
plt.plot(x,np.sin(x),color='r',label='Sin_wave')
plt.legend()
plt.subplot(2,2,2)
plt.plot(x,np.cos(x),color='g',label='Cos_wave')
plt.legend()
plt.subplot(2,2,3)
plt.plot(x,np.tan(x),color='b',label='tan_wave')
plt.legend()
plt.subplot(2,2,4)
plt.plot(x,y,color='y',label='Waveform')
plt.legend()
plt.show()
```

```python
#line plot
#scatter plot
#bar chart
#pie chart
```

```python
#scatter plot---->data is distributed(it is used to relationship between numerical values)
plt.scatter(x,np.sin(x))
```

```python
#bar chart------->
x=['orange','grapes','ban','lit']
y=[20,30,25,50]
plt.subplot(2,1,1)
plt.scatter(x,y)
plt.subplot(2,1,2)
plt.bar(x,y)
plt.show()
```

```python
#piechart
agent=['hari','swetha','Gaye','Yuva']
sales=[20,30,25,50]
plt.figure(figsize=(15,8))
plt.subplot(2,3,1)
plt.pie(sales)
plt.subplot(2,3,2)
plt.pie(sales,labels=agent)
plt.subplot(2,3,3)
plt.pie(sales,labels=agent,autopct='%.2f%%')
plt.subplot(2,3,4)
plt.pie(sales,labels=agent,autopct='%.4f%%')
plt.subplot(2,3,4)
plt.pie(sales,labels=agent,autopct='%.4f%%',explode=[0,0.2,0,0.2])
plt.show()
```
